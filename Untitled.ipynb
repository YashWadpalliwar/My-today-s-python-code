```python
def Student_Result(): 
    stu_name = input("Please enter stu name"); 
    python = int(input("Please enter python marks:- ")); 
    excel = int(input("Please enter Excel marks:- ")); 
    SQL = int(input("Please enter SQL marks:- ")); 

    # print("My python marks is",python,"Excel marks",excel); 
    Total_marks = python + excel + SQL 
    
    per = Total_marks * 100 / 300; 
    
    print(f"My python marka is :- {python} \n excel marks is {excel} and \n my sql marks is {SQL}\n My Total : {Total_marks}");
    print(f"My persentage is {per}%"); 
    
    if per > 45: 
        print("Pass"); 
    else: 
        print("Fail"); 
```

# Nested Functions
